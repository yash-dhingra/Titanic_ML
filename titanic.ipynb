```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("train.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.Age.fillna(data.Age.mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.Cabin.fillna("NA",inplace=True)
```

```python
arr=[]
for i in data.Cabin:
    if i=="NA":
        arr.append(1)
    else:
        arr.append(0)
```

```python
data["Cabin_data_missing"]=np.array(arr)
```

```python
data
```

```python
data.dtypes
```

```python
data.drop(columns="Name",axis=1,inplace=True)
```

```python
data
```

```python
dt=data
dic=dt.items()

for column,data in dic:
    if pd.api.types.is_object_dtype(data):
        dt[column]=data.astype('category').cat.as_ordered()
```

```python
dt.info()
```

```python
x,y=dt.drop(columns="Survived",axis=1),dt.Survived            
```
